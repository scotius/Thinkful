```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('metro-bike-share-trip-data.csv', low_memory=False)
```

```python
df1.head()
```

```python
starting_station_count = df1[['Starting Station ID', 'Trip ID']].groupby('Starting Station ID').count().reset_index()
```

```python
starting_station_count.rename(columns={'Starting Station ID' : 'Starting_Station_ID', 'Trip ID' : 'Trip_ID'}, inplace = True)
```

```python
starting_station_average_duration = df1['Duration'].groupby('Starting Station ID').mean()
```

```python
ending_station_count = df1[['Ending Station ID', 'Trip ID']].groupby('Ending Station ID').count().reset_index()
```

```python
df1[['Sta]].groupby('Starting Station ID').count().reset_index()
```

```python
ending_station_count.rename(columns={'Ending Station ID' : 'Ending_Station_ID', 'Trip ID' : 'Trip_ID'}, inplace = True)
```

```python
ending_station_count['Ending_Station_ID'].astype(str)
```

```python
plt.clf
plt.scatter(starting_station_count.Starting_Station_ID, starting_station_count.Trip_ID)
plt.xlabel('Starting Station')
plt.ylabel('Number of Trips')
```

```python
plt.clf
plt.scatter(ending_station_count.Starting_Station_ID, starting_station_count.Trip_ID)
plt.xlabel('Ending Station')
plt.ylabel('Number of Trips')
```
