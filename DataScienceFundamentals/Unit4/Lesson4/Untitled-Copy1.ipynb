```python
df1 = pd.read_csv('metro-bike-share-trip-data.csv', low_memory=False)
```

```python
df1.columns = df1.columns.str.replace(' ','_')
```

```python
df1 = df1.dropna()
```

```python
df1['Starting_Station_ID'] = df1.Starting_Station_ID.astype(int)
df1['Ending_Station_ID'] = df1.Ending_Station_ID.astype(int)
df1['Bike_ID'] = df1.Bike_ID.astype(int)
df1.head()
```

```python
starting_stations = df1.groupby('Starting_Station_ID')
```
